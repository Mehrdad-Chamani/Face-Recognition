# Vars

```python
# متغیرهای خالی
# مثل ایمیج ها و لیست نام ها

path = '/content/drive/MyDrive/images_pro'

# آرایه تصاویر
images = []

# نام تصاویر ( بدون پسوند )
names = []

# فهرست کردن دارکتوری های داخل مسیر پس
# py installer --> مسیردهی بایستی نسبی انجام شود
# نام فایل تمام تصاویر
myList = os.listdir(path)

print(myList)
```

```python
# فهرستی از تصاویر را بخوانیم
# انکدینگ شان را استخراج کنیم
# و با تصویر تست محک بزنیم
```

# Fetch Known Images

```python
# واکشی تصاویر و پر کردن لیست های ایمیجز و نیمز
# تصاویر را بیرون می کشیم و لیست ایمج ها و نام ها را پر می کنیم

for item in myList:
  # استخراج نام ها
  # لود با بی جی آر
  # تبدیل به آر جی بی می کنیم
  # images_pro/Ali Daei.jpg
  curImage = cv2.imread(f'{path}/{item}')

  # آرایه ایمیج ها
  images.append(curImage)

  # Ali Daei
  names.append(os.path.splitext(item)[0])

print(images)
print(names)
```

# Encode Known Images

```python
# تصاویر را انکد می کنیم
# و داخل لیستی می گذاریم

def findEncodings(images):
    encodeList = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        encodeList.append(encode)
    return encodeList
```

```python
# پنج تصویر ، هر کدام آرایه 128 اندازه ایی
# انکدینگ با یادگیری عمیق 128 اندازه را فهمیده است
# که در هر چهره چه چیزهایی بایستی اندازه گیری شود
# اندازه گیری 128 نسبت را روی چهره ها انجام داده است
# صد و بیست و هشت اندازه از جاهای خاصی از چهره
# مثلا یک نشانه خوب فاصله بین دو ابروست
# ممکن است از تصویر غلامرضا تختی باز شدگی دهانه چشم را اندازه بگیرد
known_encode_list = findEncodings(images)
print(len(known_encode_list))
print(len(known_encode_list[0]))
print(known_encode_list[0])
print("Encoding Completed!")
```

```python
# می توانیم به جای آنکه فرآیند را دستی انجام دهیم
# در دیتابیس کوئری بزنیم
# اینها رکوردهایی از دیتابیس باشند
# ردیف ها را واکشی کنیم

# حالت بهتر اینکه انکدینگ ها را هم در دیتابیس داشته باشیم
# و هر زمان لازم شد، لیست انکدینگ ها را استخراج کنیم
# و هر زمان افراد جدیدی به پایگاه داده اضافه شد
# انکدینگ ها بایستی به روز رسانی شوند

# تمام تصاویر آشنا واکشی و انکد شدند
# پویا هم هست یعنی اگر تصاویر بیشتری را ذخیره کنیم
# بدرستی عمل می کند
```

```python
# ویدئو را کپچر می کنیم
# می تواند وبکم باشد یا فیلم

cap = cv2.VideoCapture(0)
```

# Process Frames and compare them with Known Images

```python
# حلقه پردازش فریم ها

# هر فریم را می گیریم
# ممکن است یک یا چند چهره داشته باشد
# هر کدام از چهره های موجود در فریم ها را مقایسه می کنیم با
# انکدینگ های تصاویر شناخته شده
# اگر چهره ایی پیدا شد
# دور چهره یک مستطیل کشیده شود
# و بگه این فلانی است

# ممکن است دو همکار با هم وارد شرکت بشوند

# می توانیم فیلم آماده هم به حلقه پاس بدهیم

# حلقه بینهایتی که تا زمانی که کاربر کیو را نزد
# این روند ادامه پیدا می کند

# اگر چهره ایی را در تصویر پیدا کند
# انکدینگ ها را استخراج می کند

# روی تک تک فریم ها کار می کند
# تک تک چهره ها را استخراج می کنیم

# با نان لیست مقایسه می کنیم که آیا مچ هست یا نه
# اگر مچ بود مستطیلی می کشیم و می گوئیم که مثلا این فلانی است
# و در پایگاه داده ذخیره می کنیم

while True:

    # فریم --> یک شصتم ثانیه
    success, frame = cap.read()

    # کوچک می کنیم تا پردازش ها را روی آن انجام دهیم
    # چون ممکن است که تصاویر بزرگ مثلا فول اچ دی باشند
    frame_small = cv2.resize(frame, (0,0), None, 0.25, 0.25)
    fame_small = cv2.cvtColor(frame_small, cv2.COLOR_BGR2RGB)
    faces_loc = face_recognition.face_locations(frame_small)

    # چون ممکن است با چند چهره مواجه شویم
    # دو پارامتر را پاس می دهیم
    faces_encode = face_recognition.face_encodings(frame_small, faces_loc)

    # انکدفیس --> متغیر فرضی
    # انکد فیس --> چهره ایی که در این لحظه به آن فکر می کنیم
    # ممکن است که چندین چهره و لوکیشن داشته باشیم
    # zip --> حلقه دوگاه همزمان
    # به ازای هر چهره و موقعیت چهره
    # این حلقه فور داخل وایل است
    # و به ازای هر موقعیت چهره و چهره در هر فریم خاص تکرار می شود
    for encodeFace, faceLoc in zip(faces_encode, faces_loc):

        # [false, true, false, false, false]
        # [true, false, true, false, false]
        matches = face_recognition.compare_faces(known_encode_list, encodeFace)

        # این بررسی دوم را برای این می گذاریم
        # که ممکن است دو مورد تروو شود
        # ممکن است چهره تازه با دو مورد خیلی نزدیک باشد
        # خروجی یک لیست از مقادیر دیستنس ها
        face_distances = face_recognition.face_distance(known_encode_list, encodeFace)

        # اگر چند مورد تروو باشد
        # برای حصول بهترین نتیجه
        # کمترین دیستنس ها را پیدا می کنیم
        # ایندکس کمترین مقدار را بر می گرداند
        matchIndex = np.argmin(face_distances)

        # اگر تروو باشد
        # مستطیلی دور چهره می کشد
        # نام آن را هم می نویسد
        if matches[matchIndex]:

            # مثلا اگر سومین تصویر تروو باشد
            # نام آن سومین تصویر را پیدا هم پیدا می کند
            # می توانیم نام را هم ویرایش کنیم ، حروف بزرگ کنیم
            # آندر لاین ها را حذف کنیم
            name = names[matchIndex].upper()

            print(name)

            # می توانیم در سیستم حضور و غیاب
            # به جای ترسیم رکتنگل دور چهره ها
            # حضور کارکنان در پایگاه داده ثبت شود
            # به جای آنکه کارکنان دانه دانه انگشت بزنند
            # ( چون آزادی عمل را می گیرد )
            # بعضی ها ممکن است که این فکر را بکنند که
            # بایستی هر روز انگشت بزنند
            # اما این گونه سیستم یوزر فرندلی تر است
            y1,x2,y2,x1 = faceLoc

            # چون پردازش ها روی فریم اسمال انجام شده بودند
            # برای نمایش آن را بزرگتر می کنیم
            y1, x2, y2, x1 = y1*4, x2*4, y2*4, x1*4

            # ترسیم رکتنگل
            cv2.rectangle(frame,(x1,y1), (x2,y2), (255,0,255), 2)

            # رکتنگل فیل شده ، برای آنکه نام فرد را بنویسیم
            cv2.rectangle(frame, (x1,y2-35), (x2,y2), (255,0,255), cv2.FILLED)

            # نام فرد
            # با تنظیم این مختصات در مستطیل قرار می گیرد
            cv2.putText(frame,name,
                        (x1+6, y2-6), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 1)

    cv2_imshow(frame)

    # اگر کاربر دکمه کیو را زد
    # از حلقه وایل خارج می شویم
    # اگر نزد ، یک شصتم ثانیه بعد دوباره فرآیند تکرار می شود
    # و می رویم فریم بعد
    # مثلا دو همکار از جلوی دور گذر می کنند
    # و بعد همکار دیگری می آید
    # ویت ( مکث ) حلقه وایل هر یک میلی ثانیه است
    # با تغییر این ممکن است قدرت پردازشی بیشتر شود
    if cv2.waitKey(1) == ord('q'):
        break
```

```python
# آزادسازی منابع
cap.release()

# بستن پنجره ها ( مثلا کاربر با دکمه کیو از حلقه وایل خارج شود )
# اگر روی کیو زدیم پنجره بسته می شود و کار تمام است
cv2.destroyAllWindows()
```
